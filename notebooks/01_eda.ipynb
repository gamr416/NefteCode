## 1. Загрузка данных

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

DATA_PATH = 'data/'

train = pd.read_csv(os.path.join(DATA_PATH, 'daimler_mixtures_train.csv'))
test = pd.read_csv(os.path.join(DATA_PATH, 'daimler_mixtures_test.csv'))
props = pd.read_csv(os.path.join(DATA_PATH, 'daimler_component_properties.csv'))
```

## 2. Базовый обзор

```python
print("=== TRAIN ===")
print(f"Shape: {train.shape}")
print(train.columns.tolist())

print("\n=== TEST ===")
print(f"Shape: {test.shape}")
print(test.columns.tolist())

print("\n=== PROPERTIES ===")
print(f"Shape: {props.shape}")
print(props.columns.tolist())
```

## 3. Целевые переменные

```python
TARGET1 = 'Delta Kin. Viscosity KV100 - relative | - Daimler Oxidation Test (DOT), %'
TARGET2 = 'Oxidation EOT | DIN 51453 Daimler Oxidation Test (DOT), A/cm'

print(f"=== {TARGET1} ===")
print(train[TARGET1].describe())

print(f"\n=== {TARGET2} ===")
print(train[TARGET2].describe())
```

## 4. Число компонентов на сценарий

```python
components_train = train.groupby('scenario_id').size()
components_test = test.groupby('scenario_id').size()

print("TRAIN:", components_train.describe())
print("\nTEST:", components_test.describe())
```

## 5. Параметры теста

```python
temp_col = "Температура испытания | ASTM D445 Daimler Oxidation Test (DOT), °C"
time_col = "Время испытания | - Daimler Oxidation Test (DOT), ч"
bio_col = "Количество биотоплива | - Daimler Oxidation Test (DOT), % масс"
cat_col = "Дозировка катализатора, категория"

print("Температура:", sorted(train[temp_col].unique()))
print("Время:", sorted(train[time_col].unique()))
print("Биотопливо:", sorted(train[bio_col].unique()))
print("Катализатор:", sorted(train[cat_col].unique()))
```

## 6. Новые компоненты в TEST

```python
train_components = set(train['Компонент'].unique())
test_components = set(test['Компонент'].unique())
new_components = test_components - train_components
print("New components in test:", new_components)
```
